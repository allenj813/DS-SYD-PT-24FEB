# If Amazon wanted to dominate the world and build central warehouses across the world, where it the optimal place to do so?

```python
# Import libraries
from sklearn import metrics
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
import folium
import datetime as dt
import webbrowser
```

```python
# Explore the data
data = pd.read_csv('../Data/worldcities.csv')
data.head()
```

```python
print(data.shape)
print(data.describe())
print(data.info)
```

```python
print(data.isnull().sum())
```

```python
# Plot
plt.scatter(data['lat'], data['lng'], s= 20)
```

```python
df = pd.DataFrame(data, columns = ['lat', 'lng'])
print(df)
```

```python
# Plot on world map
start = [40.7128, -74.0059]

# Bild map
map = folium.Map(location=df['lat', 'lng'], zoom_start=12)
map
```

```python
data2 = data[['lat', 'lng']]

kmeans = KMeans(n_clusters = 50)
kmeans.fit(data2)
y_means = kmeans.predict(data2)
plt.scatter(data['lat'], data['lng'], c=y_means, s=20, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], c='black', s=200, alpha=0.5)
plt.xlabel('lbm')
plt.ylabel('pcBfat')
plt.show()
```
